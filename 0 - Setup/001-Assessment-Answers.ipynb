## Vocabulary:

**Python** is a programming language designed to be simple, readable, and powerful. 

Why Python? 

- There are libraries and packages that handle a wide range of programming tasks:                    
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Ex: web development, mathematical computing, machine learning models

- The python ecosystem of libraries and packages works across diverse domains.   
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Ex: In web development, we use the 'django' package (frameworks), which allows us to use JavaScript to build the interface of the website. 

- The development and maintenance of packages and libraries are community-driven. 


We can use the terminology **Libraries** and **Packages** interchangeably. There are some technical differences in the structure between them, but we are not going to worry about that. 

Main libraries we are going to focus on:              
- **pandas:** data manipulation and analysis
- **matplotlib:**: creating graphs and visualizations
- **numpy:** used for numerical and matrix operations
